```python
# Re-import necessary libraries after environment reset
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Sample data: Grades frequency
np.random.seed(42)
grades = np.random.choice(['A', 'B', 'C', 'D'], size=100, p=[0.3, 0.4, 0.2, 0.1])
grade_counts = pd.Series(grades).value_counts().sort_values(ascending=False)

# Cumulative percentage for Pareto Plot
cumulative_percentage = grade_counts.cumsum() / grade_counts.sum() * 100

# Create Rank Order & Pareto Plot
fig, ax1 = plt.subplots(figsize=(10, 6))

# Bar plot (Rank Order)
ax1.bar(grade_counts.index, grade_counts.values, color='skyblue', label='Frequency')
ax1.set_ylabel('Frequency')
ax1.set_title('Pareto Plot (Rank Order + Cumulative %)')
ax1.set_xlabel('Grade')

# Line plot (Cumulative %)
ax2 = ax1.twinx()
ax2.plot(grade_counts.index, cumulative_percentage, color='red', marker='o', label='Cumulative %')
ax2.set_ylabel('Cumulative Percentage (%)')
ax2.set_ylim(0, 110)

# Combine legends from both axes
lines, labels = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax1.legend(lines + lines2, labels + labels2, loc='upper left')

plt.grid(True)
plt.show()
```
