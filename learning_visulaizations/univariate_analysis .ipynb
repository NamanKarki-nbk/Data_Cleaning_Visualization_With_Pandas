```python
# Re-importing necessary libraries after reset
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

# Create a sample dataset
np.random.seed(42)
data = pd.DataFrame({
    'Gender': np.random.choice(['Male', 'Female'], size=100),
    'Score': np.random.normal(loc=75, scale=10, size=100).astype(int),
    'Grade': np.random.choice(['A', 'B', 'C', 'D'], size=100, p=[0.3, 0.4, 0.2, 0.1])
})

# Count of Grades for bar and pie charts
grade_counts = data['Grade'].value_counts()

# Set up 2x2 plot layout
fig, axs = plt.subplots(2, 2, figsize=(14, 10))

# Bar Chart
axs[0, 0].bar(grade_counts.index, grade_counts.values, color='skyblue')
axs[0, 0].set_title("Bar Chart - Count of Grades")
axs[0, 0].set_xlabel("Grade")
axs[0, 0].set_ylabel("Count")

# Pie Chart
axs[0, 1].pie(grade_counts.values, labels=grade_counts.index, autopct='%1.1f%%', startangle=90)
axs[0, 1].set_title("Pie Chart - Grade Distribution")

# Histogram
sns.histplot(data['Score'], bins=10, kde=True, ax=axs[1, 0], color='orange')
axs[1, 0].set_title("Histogram - Score Distribution")
axs[1, 0].set_xlabel("Score")
axs[1, 0].set_ylabel("Frequency")

# Jitter/Dot Plot
sns.stripplot(data=data, x='Gender', y='Score', jitter=True, ax=axs[1, 1])
axs[1, 1].set_title("Jitter/Dot Plot - Scores by Gender")
axs[1, 1].set_xlabel("Gender")
axs[1, 1].set_ylabel("Score")

plt.tight_layout()
plt.show()
```
